```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

from selenium import webdriver
import chromedriver_autoinstaller
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()
```

```python
browser.get("https://shopee.co.id/shop/40847197/recommendation-landing?upstream=")
```

```python
soup = BeautifulSoup(browser.page_source,'lxml')
```

```python
row = soup.find_all('div', class_='_9cGCjS')
```

```python
data = []

for row in row:
    nama_produk_elem = row.find(class_='_89o+uF mC2b-a d-wZla')
    harga_elem = row.find(class_='_1cMhqd waO4ig P4VCSA')
    diskon_elem = row.find(class_='percent')
    harga_diskon_elem = row.find(class_='_1cMhqd tdQs-h')
    terjual_elem = row.find(class_='r4l-mr R9rrSp')
    kota_elem = row.find(class_='qR3ZfT')

    nama_produk = nama_produk_elem.text if nama_produk_elem else None
    harga = harga_elem.text if harga_elem else None
    diskon = diskon_elem.text if diskon_elem else None
    harga_diskon = harga_diskon_elem.text if harga_diskon_elem else None
    terjual = terjual_elem.text if terjual_elem else None
    kota = kota_elem.text if kota_elem else None

    dt = {
        'nama_produk': nama_produk,
        'harga': harga,
        'diskon': diskon,
        'harga_diskon': harga_diskon,
        'terjual': terjual,
        'kota': kota
    }
    data.append(dt)
```

```python
df = pd.DataFrame(data)
```
